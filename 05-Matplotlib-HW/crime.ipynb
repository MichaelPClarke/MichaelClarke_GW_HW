```python
# Repeat plotting scatter and linear model for murder rate versus year

x_values1 = crime_data['year']
y_values1 = crime_data.iloc[:,5]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values1, y_values)
regress_values = x_values1 * slope + intercept
# Create equation of line to calculate predicted violent crime rate
line_eq1 = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plot the linear model on top of scatter plot 
plt.scatter(x_values1,y_values)
plt.xticks(x_values1, rotation=90)
plt.plot(x_values1,regress_values,"r-")
plt.annotate(line_eq1,(2005, 8),fontsize=12,color="red")
plt.xlabel('Year')
plt.ylabel('Murder Rate')
plt.show()
```

```python
# Repeat plotting scatter and linear model for aggravated assault versus year
x_values2 = crime_data['year']
y_values2 = crime_data['aggravated assault rate']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values2, y_values)
regress_values = x_values2 * slope + intercept
# Create equation of line to calculate predicted violent crime rate
line_eq2 = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plot the linear model on top of scatter plot 
plt.scatter(x_values2,y_values)
plt.plot(x_values2,regress_values,"r-")
plt.xticks(x_values2, rotation=90)
plt.annotate(line_eq2,(2003,350),fontsize=12,color="red")
plt.xlabel('Year')
plt.ylabel('Aggravated Assault Rate')
plt.show()
```

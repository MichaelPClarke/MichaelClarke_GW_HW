```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file
file_df = pd.read_csv(file, low_memory=False)
file_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc
columns_df = file_df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
columns_df
```

```python
# Change reading 0 and 1 to No and Yes, respectively
cleaned_df = columns_df.replace(to_replace=[0, 1], value=["No","Yes"])
cleaned_df
```
